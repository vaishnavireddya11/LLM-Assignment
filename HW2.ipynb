```python
from transformers import pipeline
generator = pipeline("text-generation", model="gpt2")
print(generator("AI will revolutionize", max_length=30)) #text-generation
```

```python
summarizer = pipeline("summarization", model="google/pegasus-xsum")
result = summarizer("Artificial Intelligence is transforming the world of education and jobs...", max_length=30)
print(result) #summarization
```

```python
qa = pipeline("question-answering", model="google/flan-t5-large")
result = qa(question="Who invented AI?", context="AI was invented by various researchers including Alan Turing.")
print(result) #question and answering
```

```python
translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
print(translator("Artificial Intelligence is powerful.", max_length=40)) #translation
```
